モデルの定義だけ  
pytorch_forcastingというライブラリを見つけた。

```python
import torch
from pytorch_forecasting.models.temporal_fusion_transformer import TemporalFusionTransformer
import torchmetrics

# Temporal Fusion Transformer モデルの初期化
model = TemporalFusionTransformer(hidden_size=32, 
                                  lstm_layers=2, 
                                  dropout=0.1,
                                  #loss=torch.nn.CrossEntropyLoss(),
                                  logging_metrics=[],
                                  loss=torchmetrics.MeanSquaredError(), # torchmetricsのMSEを使用
                                  output_size=8)  # 8クラス分類の設定

# モデルの構造を表示
print(model)
```
